#### Linear Regression (Only Least Squares) Model
- Import Libraries
- Import CSV as pandas dataframe
- Feature Engineering (Dummy Encoding)
- Splitting Data into train test 
- Fit Data to OLS Model (Linear Regression Model)
- Performance Analysis

```python
#import required packages
import numpy as np
import pandas as pd
#Train Test Split Package
from sklearn.model_selection import train_test_split
#Import statsmodels module
import statsmodels.api as sm
```

```python
#import CSV fill as Pandas Dataframe
df = pd.read_csv("supermarket_sales - Sheet1.csv")
#Create a copy of the original dataframe
df_origial = df.copy(deep=True)
```

```python
#Dummy Encoding the Data
data_encoded = pd.get_dummies(df[['City','Customer type','Gender','Unit price','Product line','Quantity','Payment','cogs','Rating','Total','gross income']], drop_first = True)
data_encoded.head()
```

##### Splitting Data into Train Test Partitions

```python
#Performing train test split on the data
train, test = train_test_split(data_encoded,test_size = 0.3,random_state=31)
X_train = train.drop(['Total'],axis=1)
y_train = train.filter(['Total'])
X_test = test.drop(['Total'],axis=1)
y_test = test.filter(['Total'])
```

##### OLS Regression Model

```python
#Initialize the OLS model
##Training the OLS model on segregated training data
model = sm.OLS(y_train,X_train)
```

```python
#Testing the OLS model on segragated test data
results = model.fit()
```

##### OLS Model Performance Anaysis

```python
#Display Intercept
display(results.params[0])
#Display Intercept
display(results.params[1])

#Extracting other metrics
print('Parameters: ', results.params)
print('Standard errors: ', results.bse)
print('Predicted values: ', results.predict())
```

```python
#Summary Report
print(results.summary())
```
